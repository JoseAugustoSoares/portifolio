```python
import sys
#!{sys.executable} -m pip install opencv-python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
from skimage import io, data
```

```python
#Passo 1: Aquisio da Imagem
pathimagem = 'drone3.jpg'
imgColorida = cv2.imread(pathimagem)
plt.imshow(imgColorida)
```

```python
#Passo 2: Converso para tons de cinza se a foto inicial for colorida
img = cv2.cvtColor(imgColorida, cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap='gray')
```

```python
#Passo 3: Blur/Suavizao da imagem
suave = cv2.blur(img, (3, 8))
plt.imshow(suave, cmap='gray')
```

```python
#Passo 4: Defina o tamanho do kernel para a operao de abertura
kernel_size = 5
kernel = np.ones((kernel_size, kernel_size), np.uint8)
opened_image = cv2.morphologyEx(suave, cv2.MORPH_OPEN, kernel)
plt.imshow(opened_image, cmap='gray')
```

```python
#Passo 5: Binarizando a imagem resultando em pixels brancos e pretos
ret, bina = cv2.threshold(opened_image, 195, 255, 0)
plt.imshow(bina, cmap='gray')
```

```python
#Passo 6: Deteco de bordas por Canny
bordas = cv2.Canny(bina, 70, 150)   # 70,150
plt.imshow(bordas, cmap='gray')
```

```python
#Passo 7: Identificao dos contornos da imagem.
objetos, hierarchy = cv2.findContours(bordas, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```python
print("Objetos Detectados na imagem:",len(objetos))
```
